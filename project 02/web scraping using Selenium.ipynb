```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the WebDriver for Microsoft Edge
driver = webdriver.Edge()

# Define the search query
query = "watch"

# Navigate to the Flipkart search page for watches
driver.get(f"https://www.flipkart.com/search?q={query}watch&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")

# Wait for page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'hCKiGj')))

# Find all product items based on their class name
elems = driver.find_elements(By.CLASS_NAME, 'hCKiGj') 

# Create lists to hold the extracted data
product_names = []
product_prices = []
product_discounts = []
product_details = []

# Loop through and extract the text of each element (product)
for elem in elems:
    details = elem.text.split('\n')
    product_names.append(details[0])  # Product name
    product_prices.append(details[1])  # Product price
    product_discounts.append(details[2])  # Product discount
    product_details.append('\n'.join(details[3:]))  # Other details

# Create a DataFrame using the extracted data
data = {
    'Product Name': product_names,
    'Price': product_prices,
    'Discount': product_discounts,
    'Details': product_details
}
df = pd.DataFrame(data)

# Save the DataFrame to an Excel file
df.to_excel('watches_data.xlsx', index=False)

# Close the browser after the operation
driver.quit()
```
